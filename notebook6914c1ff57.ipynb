Exploratory Data Analysis

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/kaggle/input/house-prices-advanced-regression-techniques/train.csv')
```

```python
data.head()
```

```python
data.columns
```

```python
data.shape
```

```python
missing=data.isnull().sum()
missing=missing[missing>0]
missing.sort_values(inplace=True)
missing.plot.bar()
```

```python
data['SalePrice'].describe()
```

```python
sns.set(rc={'figure.figsize':(12,8)})
sns.distplot(data['SalePrice'],kde=False,bins=20);
```

```python
sns.kdeplot(data['SalePrice'])
```

```python
sns.scatterplot(x='GarageArea',y='SalePrice',data=data)
sns.scatterplot(x='OverallQual',y='SalePrice',data=data)
sns.scatterplot(x='FullBath',y='SalePrice',data=data)
```

```python
sns.regplot(x='GarageCars',y='GarageArea',data=data,scatter=True,fit_reg=True)
```

```python
sns.boxplot(x=data["SalePrice"])
```

```python
f,ax=plt.subplots(figsize=(19,8))
fig=sns.boxplot(x='SaleType',y='SalePrice',data=data)
fig.axis(ymin=0,ymax=800000);
xt=plt.xticks(rotation=45)
```

```python
f,ax=plt.subplots(figsize=(19,8))
fig=sns.boxplot(x='OverallQual',y='SalePrice',data=data)
fig.axis(ymin=0,ymax=800000);
```
