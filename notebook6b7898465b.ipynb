```python
import torch
from diffusers import StableDiffusionPipeline
from IPython.display import display
```

```python
model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe.to("cuda")

prompt = "civlisation on moon in ultrarealistic way with man driving cars and stuff"
image = pipe(prompt, num_inference_steps=100, guidance_scale=8.5).images[0]
display(image)
```

```python
image.save("ai_image.png")
print("AI Image has been generated and saved as ai_image.png")
```
